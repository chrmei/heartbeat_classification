```python
import numpy as np
import pandas as pd
from lazypredict.Supervised import LazyClassifier
```

```python
#import PTB data
X_ptb_train_sm = pd.read_csv('../data/PTB/X_ptb_train_sm.csv')
y_ptb_train_sm = pd.read_csv('../data/PTB/y_ptb_train_sm.csv')

X_ptb_val = pd.read_csv('../data/PTB/X_ptb_val.csv')
y_ptb_val = pd.read_csv('../data/PTB/y_ptb_val.csv')

X_ptb_test = pd.read_csv('../data/PTB/X_ptb_test.csv')
y_ptb_test = pd.read_csv('../data/PTB/y_ptb_test.csv')
```

```python
#LazyClassifier on PTB SMOTE training data, test data
clf = LazyClassifier(verbose=1,ignore_warnings=True, custom_metric=None)

models,predictions = clf.fit(X_ptb_train_sm, X_ptb_test, y_ptb_train_sm, y_ptb_test)
```
