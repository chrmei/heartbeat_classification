```python
#Statistical test to prove class imbalance

#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chisquare.html
#https://en.wikipedia.org/wiki/Goodness_of_fit

#H0: balanced class distribution -> each class identical frequency: MIT 109446 rows/5 classes = 21889.2, PTB 14545 rows/2 classes = 7272.5
#H1: unbalanced class distribution

from scipy.stats import chisquare


#MIT 
class_mit = df_mitbih[187]  # column of combined MIT dataset with class annotations

# Step 1: Count observed frequencies
class_counts_mit = class_mit.value_counts().sort_index()
observed_mit = class_counts_mit.values
num_classes_mit = len(observed_mit)
total_mit = observed_mit.sum()

# Step 2: Expected counts (uniform distribution)
expected_mit = total_mit / num_classes_mit

# Step 3: Run Chi-square goodness-of-fit
chi2_mit, p_mit = chisquare(f_obs=observed_mit, f_exp=expected_mit)

print("Observed:", observed_mit)
print("Expected:", expected_mit)
print("Chi-square:", chi2_mit, "p-value:", p_mit)

print('p<0.05 -> H0 is rejected, H1: unbalanced class distribution for MIT dataframe.')
print('---------------------------------------------------------------------------')



#PTB
class_ptb = df_ptbdb[187]  # column of combined PTB dataset with class annotations

# Step 1: Count observed frequencies
class_counts_ptb = class_ptb.value_counts().sort_index()
observed_ptb = class_counts_ptb.values
num_classes_ptb = len(observed_ptb)
total_ptb = observed_ptb.sum()

# Step 2: Expected counts (uniform distribution)
expected_ptb = total_ptb / num_classes_ptb

# Step 3: Run Chi-square goodness-of-fit
chi2_ptb, p_ptb = chisquare(f_obs=observed_ptb, f_exp=expected_ptb)

print("Observed:", observed_ptb)
print("Expected:", expected_ptb)
print("Chi-square:", chi2_ptb, "p-value:", p_ptb)

print('p<0.05 -> H0 is rejected, H1: unbalanced class distribution for PTB dataframe.')
print('---------------------------------------------------------------------------')
```
